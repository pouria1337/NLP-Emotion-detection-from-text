## Delete rows whose Field1 does not correspond with NRC lexicon

```python
# Mount drive and create a dataframe of the NRC Lexicon
from google.colab import drive
drive.mount('/content/drive/')
nrc_df = pd.read_csv('/content/drive/My Drive/NRC.txt', encoding= 'unicode-escape', sep="\t")
nrc_df.columns = ['Word', 'Emotion', 'Member']
```

```python
nrc_df
```

```python
# Get uniqe values of nrc_df and df (isear data)
set1 = set(nrc_df["Emotion"].unique())
set2 = set(df["Field1"].unique())

mutual_values = set1.intersection(set2)
print(mutual_values)
```

We will clean both datasets (isear and nrc) according to the mutual values

```python
# Clean ISEAR dataset according to the mututal values
df2 = df[df["Field1"].isin(mutual_values)]
df2 = df2.reset_index(drop=True)
df2
```

```python
# Clean NRC according to the mutual values
nrc_df2 = nrc_df[nrc_df["Emotion"].isin(mutual_values)]
nrc_df2 = nrc_df2.reset_index(drop=True)
nrc_df2
```

```python
# Keep only the rows from NRC dataframe whose member column is 1
condition = nrc_df2["Member"] == 1
nrc_df3 = nrc_df2[condition]
nrc_df3 = nrc_df3.reset_index(drop=True)
nrc_df3
```

```python
# We now try to represent each emotion by a list of words in one dictionary
emowords = dict()
def create_dict(emotion):
    temp = nrc_df3[nrc_df3["Emotion"] == emotion]["Word"]
    temp = list(temp)
    emowords[emotion] = temp
```

```python
# Fill emowords with emotions
for i in mutual_values:
    create_dict(i)
emowords
```

## Calculate cosine similarity

```python
# Now, we represent each emotion by a group of words
# For every memory in ISEAR, we calculate its proximity or similarity with each
# group of words for every emotion
```

```python
import torch
!pip install sentence-transformers
from sentence_transformers import SentenceTransformer, util
model = SentenceTransformer('all-mpnet-base-v2')
```

```python
query_embedding = model.encode("London's population")
passage_embedding = model.encode(['London has 9,787,426 inhabitants at the 2011 census',
                                  'London is known for its finacial district',
                                 'Hello, Tehran is the capital of Iran!',
                                  'Britain is a very large kingdom.',
                                  'London is a big metropolice'])

x = util.cos_sim(query_embedding, passage_embedding)
x.shape
print("Similarity:", x)
```

## Encode emotion words to numpy and then to Torch

```python
emoword_torch = dict()
def torch_emowords(emotion):
    emoword_torch[emotion] = torch.from_numpy(model.encode(emowords[emotion]))

for i in mutual_values:
    torch_emowords(i)
emoword_torch
```

```python
matrix_01 = list()
for memory in df2['SIT']:
  query_embedded_tch = torch.from_numpy(model.encode(memory))
  for key in emoword_torch:
    matrix_01.append(torch.mean(util.cos_sim(query_embedded_tch, emoword_torch[key])))
```

```python
len(matrix_01), len(df2)
```

```python
# Create a corresponding 2d matrix
# First, convert from tensor to numpy for better compatibility with scikit-learn
import numpy as np
matrix_numpy = []
matrix_01_copy = matrix_01.copy()
for tensor in matrix_01_copy:
    matrix_numpy.append(tensor.numpy())

#Changing the matrix dtype to numpy
matrix_numpy = np.array(matrix_numpy)
type(matrix_numpy)

matrix_numpy_1d = np.ravel(matrix_numpy)


matrix_2d_alpha = matrix_numpy_1d.reshape((len(df2),len(mutual_values)))
matrix_2d_alpha
```

```python
matrix_df = pd.DataFrame(matrix_2d_alpha, columns=['anger', 'disgust', 'fear', 'joy', 'sadness'])
matrix_df['Recorded Emotion'] = df2['Field1']
matrix_df
```

```python
matrix_df.to_csv("/content/similarities.csv", index = False)
```

## Apply machine learning classification on the calculated similarities

```python
import pandas as pd
import numpy as np
```

```python
matrix_df = pd.read_csv("/content/similarities.csv")
```

```python
matrix_b = matrix_df.copy()
#Run Random Forest
from sklearn.model_selection import train_test_split

features = list(matrix_b.columns)[:-1]
labels = list.pop(list(matrix_b.columns))

#splitting features and lables
X = matrix_b[features]
y = matrix_b[labels]

X_np = np.array(X)
y_np = np.array(y).ravel()

#Creating train/test splits
X_train, X_test, y_train, y_test = train_test_split(X_np, y_np, test_size=0.2, random_state=42)
len(X_train), len(X_test), len(y_train), len(y_test)
```

```python
# Import necessary modules for classification
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier, RandomForestClassifier
from sklearn import metrics, svm
from sklearn.metrics import make_scorer, classification_report
from sklearn.model_selection import cross_validate, cross_val_score, cross_val_predict, KFold
from sklearn.datasets import make_classification
from sklearn.naive_bayes import GaussianNB
```

```python
# Function to get metrics from resulting dictionary and round them to 3
def getScores(dict):
  scores = ['test_accuracy', 'test_f1_macro', 'test_precision_macro', 'test_recall_macro']
  temp_ls= []
  for score in scores:
    temp = 0
    temp = round(dict[score].mean(), 3)
    temp_ls.append(temp)
  return temp_ls
```

```python
# Logistic Regression classification
clf_LoReg = LogisticRegression(random_state=33).fit(X,y)
clf_LoReg.score(X,y)
LR_cv = cross_validate(clf_LoReg, X_np, y_np, cv=10, scoring=['accuracy', 'precision_macro', 'recall_macro', 'f1_macro'])
w = getScores(LR_cv)
w
```

```python
# Training RF
rf_clf = RandomForestClassifier(n_estimators=100)
rf_cv_macro = cross_validate(rf_clf, X_np, y_np, cv=10,scoring=['accuracy', 'precision_macro', 'recall_macro', 'f1_macro'])
x = getScores(rf_cv_macro)
x
```

```python
# SVM Classification
from sklearn import svm
svm_clf = svm.SVC(kernel='rbf', C=1, random_state=42)
svm_cv = cross_validate(svm_clf, X_np, y_np, cv=10, scoring=['accuracy', 'precision_macro', 'recall_macro', 'f1_macro'])
y = getScores(svm_cv)
y
```

```python
# Gradient Boosting classification
gb_cv = cross_validate((GradientBoostingClassifier(
                        n_estimators=100, learning_rate=1.0, max_depth=1, random_state=0)),X_np, y_np, cv=10,
                        scoring=['accuracy', 'precision_macro', 'recall_macro', 'f1_macro'])
z = getScores(gb_cv)
z
```

```python
# Create a dataframe for all classifications
df_1 = pd.DataFrame(index=['Accuracy','F1-score','Precision', 'Recall'], columns=['Random_Forest', 'SVM', 'G-Boosted_Tree'])
df_1['Random_Forest'] = x
df_1['SVM'] = y
df_1['G-Boosted_Tree'] = z
df_1
```

## Metrics for each emotion for each classifier

```python
# Create the dataframe for it
iterables = [["anger", "disgust", "fear", "joy", "sadness"], ["accuracy", "f1-score", "precision", "recall"]]
index = pd.MultiIndex.from_product(iterables, names=["Emotion", "Metrics"])
df_2 = pd.DataFrame(index = index, columns=['(Random_Forest)', '(SVM)', '(G-Boosted_Tree)', '(NaiveBayes_GaussianNB)'])
```

```python
y_pred_gb = cross_val_predict(GradientBoostingClassifier(n_estimators=100, learning_rate=0.9, max_depth=1, random_state=0), X_np, y_np, cv=10)
gb_class_report = classification_report(y_np, y_pred_gb, output_dict=True)

y_pred_nbg = cross_val_predict(GaussianNB(), X_np, y_np, cv=10)
nbg_class_report = classification_report(y_np, y_pred_nbg, output_dict=True)

y_pred_rf = cross_val_predict(rf_clf, X_np, y_np, cv=10)
rf_class_report = classification_report(y_np, y_pred_rf, output_dict=True)

y_pred_svm = cross_val_predict(svm_clf, X_np, y_np, cv=10)
svm_class_report = classification_report(y_np, y_pred_svm,output_dict=True)
```

```python
list1 = ['anger', 'disgust', 'fear', 'joy', 'sadness']
list2 = ['f1-score', 'precision', 'recall']

#FOR RANDOM--FOREST
for i in list1:
  for j in list2:
    df_2['(Random_Forest)'][i][j] = round(rf_class_report[i][j], ndigits=2)
    df_2['(Random_Forest)'][i]['accuracy'] = round(rf_class_report['accuracy'], ndigits=2)

#FOR SVM
for i in list1:
  for j in list2:
    df_2['(SVM)'][i][j] = round(svm_class_report[i][j], ndigits=2)
    df_2['(SVM)'][i]['accuracy'] = round(svm_class_report['accuracy'], ndigits=2)


#(G-Boosted_Tree)
for i in list1:
  for j in list2:
    df_2['(G-Boosted_Tree)'][i][j] = round(gb_class_report[i][j], ndigits=2)
    df_2['(G-Boosted_Tree)'][i]['accuracy'] = round(gb_class_report['accuracy'], ndigits=2)


#(NaiveBayes_GaussianNB)
for i in list1:
  for j in list2:
    df_2['(NaiveBayes_GaussianNB)'][i][j] = round(nbg_class_report[i][j], ndigits=2)
    df_2['(NaiveBayes_GaussianNB)'][i]['accuracy'] = round(nbg_class_report['accuracy'], ndigits=2)
```
